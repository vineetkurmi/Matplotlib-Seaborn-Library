## Part C - Matplotlib

### Q31. Create a Bar Chart of Category-wise Sales.

```python
import matplotlib.pyplot as plt
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

### Q32. Create a Pie Chart of Segment distribution.

```python
df.groupby('Segment')['Sales'].sum().plot(kind='pie')
plt.title('Segment distribution')
plt.show()
```

### Q33. Create a Histogram of Sales.

```python
df['Sales'].plot(kind='hist')
plt.title('Sales')
plt.show()
```

### Q34. Create a Line Chart of monthly Sales.

```python
df.groupby(df['Order Date'].dt.month)['Sales'].sum().plot(kind='line')
plt.title('Monthly Sales')
plt.show()
```

### Q35. Create a Scatter Plot of Sales vs Profit.

```python
plt.scatter(df['Sales'],df['Profit'])
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

### Q36. Create a Box Plot of Profit.

```python
plt.boxplot(df['Profit'])
plt.title('Profit')
plt.show()
```

### Q37. Create a Horizontal Bar Chart of Top 10 States by Sales.

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 States by Sales')
plt.gca().invert_yaxis()
plt.show()
```

## Part D - Seaborn

### Q38. Create a Count Plot of Category.

```python
import seaborn as sns
sns.countplot(data=df , x='Category')
plt.title('Category')
plt.show()
```

### Q39. Create a Bar Plot of Profit by Region.

```python
sns.barplot(data=df , x='Region' , y='Profit')
plt.title('Profit by Region')
plt.show()
```

### Q40. Create a Box Plot of Sales by Segment.

```python
sns.boxplot(data=df , x='Segment' , y='Sales')

plt.title('Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Sales')

plt.show()
```

### Q41. Create a Strip Plot of Sales by Category.

```python
sns.stripplot(data=df , x='Category' , y='Sales')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```
